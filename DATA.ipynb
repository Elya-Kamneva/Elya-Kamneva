```python
#Заработная плата на 2 получателя для Postman
import json

# Настройки для генерации
iterations = 20
start_amount = 101
start_number = 5656

# Пустой список для хранения всех итераций
data = []

for i in range(iterations):
    amount = start_amount + i  # Счетчик для поля amount
    number = start_number + i  # Счетчик для номера документа
    
    # Для простоты разделим amount на 2 части
    employee_amount_1 = int(amount * 0.5)  # Половина от суммы
    employee_amount_2 = amount - employee_amount_1  # Остаток
    
    # Формируем структуру для текущей итерации
    iteration_data = {
        "amount": amount,
        "number": number,
        "employee_amount_1": employee_amount_1,
        "employee_amount_2": employee_amount_2,
    }
    
    # Добавляем данные в список
    data.append(iteration_data)

# Сохраняем данные в JSON файл
with open("output.json", "w", encoding="utf-8") as json_file:
    json.dump(data, json_file, ensure_ascii=False, indent=4)

#print(data)
print("JSON файл успешно создан!")
```
