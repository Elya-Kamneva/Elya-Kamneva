```python
#На ферме есть куры, коровы и свиньи. У курицы две ноги, у свиньи - четыре, у коровы - четыре
#Задача: Запросить у пользователя (фермера) ввод кол-ва кур, свиней и коров, просуммировать кол-во ног всех животных и вывести результат на консоль.

print('Добро пожаловать на ферму! Давайте посчитаем количество ног на ферме?')

сhicken_legs = 2                                                    #Задаем количество ног у животных
pig_legs = 4
cow_legs = 4

сhicken_input = input('Сколько куриц на ферме? ')                   #Запрашиваем количество куриц на ферме
сhicken = int(сhicken_input)                                        #Переводим введенное значение в число

pig_input = input('Сколько свиней на ферме? ')                      #Запрашиваем количество свиней на ферме
pig = int(pig_input)                                                #Переводим введенное значение в число

cow_input = input('Сколько коров на ферме? ')                       #Запрашиваем количество коров на ферме
cow = int(cow_input)                                                #Переводим введенное значение в число

all_legs = сhicken*сhicken_legs + pig*pig_legs + cow*cow_legs       #Считаем общее количество ног на ферме

print("{} {} {}".format('Ого! На ферме всего',all_legs,'ног!'))     #Выводим результат
```
